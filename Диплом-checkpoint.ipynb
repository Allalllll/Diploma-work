```python
import matplotlib.pyplot as plt
import seaborn as sns

#  информация о данных
print("Информация о данных пользователей:")
print(users_df.info())

# описание данных
print("\nСтатистическое описание данных:")
print(users_df.describe(include='all'))

# есть ли пропуски?
print("\nКоличество пропусков в данных:")
print(users_df.isnull().sum())

# возраст пользователей
plt.figure(figsize=(10, 5))
sns.histplot(users_df['age'], bins=20, kde=True, color='blue')
plt.title('Распределение возраста пользователей')
plt.xlabel('Возраст')
plt.ylabel('Количество')
plt.show()

# пол пользователей
sex_counts = users_df['sex'].value_counts()
plt.figure(figsize=(8, 5))
sns.barplot(x=sex_counts.index, y=sex_counts.values, palette='viridis')
plt.title('Распределение по полу')
plt.xlabel('Пол (1 - Мужчины, 2 - Женщины, 0 - Не указан)')
plt.ylabel('Количество')
plt.show()

# города пользователей (топ 10)
top_cities = users_df['city_id'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_cities.index, y=top_cities.values, palette='plasma')
plt.title('Топ 10 городов по количеству пользователей')
plt.xlabel('ID города')
plt.ylabel('Количество')
plt.show()
```

```python
# загрузка данных
history_df = pd.read_csv('history.tsv', sep='\t')

# информация о данных
print("Информация о данных истории:")
print(history_df.info())

# описание данных
print("\nСтатистическое описание данных:")
print(history_df.describe(include='all'))

# проверка пропусков
print("\nКоличество пропусков в данных:")
print(history_df.isnull().sum())

# распределение ставок CPM
plt.figure(figsize=(10, 5))
sns.histplot(history_df['cpm'], bins=30, kde=True, color='green')
plt.title('Распределение ставок CPM')
plt.xlabel('CPM')
plt.ylabel('Количество')
plt.show()

# 5. Анализ активности по часам
hour_counts = history_df['hour'].value_counts().sort_index()
plt.figure(figsize=(10, 5))
sns.barplot(x=hour_counts.index, y=hour_counts.values, palette='coolwarm')
plt.title('Активность по часам')
plt.xlabel('Час')
plt.ylabel('Количество показов')
plt.show()

# топ площадок
top_publishers = history_df['publisher'].value_counts().head(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_publishers.index, y=top_publishers.values, palette='mako')
plt.title('Топ-10 площадок по количеству показов')
plt.xlabel('Площадка')
plt.ylabel('Количество показов')
plt.show()
```

```python
# корреляция
correlation_matrix = history_df.corr()

# визуализация корреляции 
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Корреляционная матрица")
plt.show()
```

```python
# группировка данных по столбцу 'hour' и подсчет количества показов
hourly_activity = history_df['hour'].value_counts().sort_index()

# 
plt.figure(figsize=(12, 6))
sns.barplot(x=hourly_activity.index, y=hourly_activity.values, palette='viridis')
plt.title('Активность пользователей по часам')
plt.xlabel('Час')
plt.ylabel('Количество показов')
plt.xticks(rotation=45)
plt.show()
```

```python
# Группировка по часам дня
hourly_activity_by_day = history_df.groupby(history_df['hour'] % 24)['hour'].count()

# Построение графика
plt.figure(figsize=(14, 7))
sns.barplot(x=hourly_activity_by_day.index, y=hourly_activity_by_day.values, palette='viridis')
plt.title('Активность пользователей по времени суток (0–23)', fontsize=16)
plt.xlabel('Час суток', fontsize=14)
plt.ylabel('Количество показов', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

```python
# Удаление выбросов: возраста больше 100
users_df = users_df[users_df['age'] <= 100]

# Добавление категории для пользователей с возрастом 0
users_df['age_category'] = users_df['age'].apply(
    lambda age: 'Возраст не указан' if age == 0 else (
        'Молодежь' if age <= 25 else (
            'Взрослые' if 26 <= age <= 55 else 'Пожилые'
        )
    )
)

# Проверка распределения возрастных категорий
print(users_df['age_category'].value_counts())
```

```python

# Подсчет количества пользователей в каждой категории
age_counts = users_df['age_category'].value_counts()

# Построение графика
plt.figure(figsize=(8, 5))
sns.barplot(x=age_counts.index, y=age_counts.values, palette='viridis')

plt.xlabel('Возрастная категория', fontsize=14)
plt.ylabel('Количество пользователей', fontsize=14)
plt.xticks(fontsize=12, rotation=15)
plt.yticks(fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Подсчет количества пользователей по полу
sex_counts = users_df['sex'].value_counts()

# Замена числовых значений на текстовые категории
sex_labels = {1: 'Мужчины', 2: 'Женщины', 0: 'Не указан'}
sex_counts.index = sex_counts.index.map(sex_labels)

# Расчет процентного соотношения
sex_percent = (sex_counts / sex_counts.sum()) * 100

# Построение круговой диаграммы
plt.figure(figsize=(8, 8))
plt.pie(
    sex_percent, 
    labels=sex_counts.index, 
    autopct='%1.1f%%', 
    startangle=90, 
    colors=['#4DAF4A', '#377EB8', '#FF7F00'], 
    textprops={'fontsize': 12}
)
plt.title('Процентное соотношение пользователей по полу', fontsize=16)
plt.show()
```
